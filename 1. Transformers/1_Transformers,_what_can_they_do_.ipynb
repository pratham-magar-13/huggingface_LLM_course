**Transformers** are used to solve all kinds of tasks across different modalities, including NLP, computer vision, audio processing and more.

The most basic object in the transformers library is the pipeline() function.

It connects a model with its necessary preprocessing and postprocessing steps, allowing us to directly input any text and get an intelligible answer

# Example of a sentiment analysis

It performs text classificaiton on given input, and determines if its positive or negative.


```python
from transformers import pipeline
classifier=pipeline("sentiment-analysis")
classifier("I've been waiting for Manchester United to play in the Champions Leage")
```

```python
classifier(['United are a loser team.','But, I hope they do better next season.'])
```

# Example of zero shot classification pipeline
It is more general text-classification pipeline: it allows us to provide the labels we want

```python
classifier2=pipeline('zero-shot-classification')
classifier2(
    'Laliga is the 2nd best league',
    candidate_labels=['Sports','Politics','Education'],
)
```

 Here, sports has the highest confidence of 0.8 than others

# Text generation example

Output is generated with a bit of randomness. It changes each time we call the generator object.

```python
generator=pipeline('text-generation')
generator('Football is the')
```

```python
# Loading another model for text generation
generator2=pipeline('text-generation',model='distilgpt2')
generator2('Football is the',max_length=30,num_return_sequences=2)
```

# Mask filling

to fill in the blanks

```python
unmasker = pipeline("fill-mask")
unmasker("This course will teach you all about <mask> models.", top_k=2)
```

# Named Entity Recognition(NER)
task where the model has to find which parts of the input text correspond to entities such as persons, location, or organiztions.

```python
from transformers import pipeline
ner = pipeline("ner", aggregation_strategy="simple")
ner("My name is John and I work at Hugging Face in Brooklyn.")
```

# Question Answering

Answers questions using information from a given context.

```python
question_answerer = pipeline("question-answering")
question_answerer(
    question=["What is my name?", "Where do I work?"],
    context="My name is John and I work at Hugging Face in Brooklyn",
)
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

# Load the tokenizer and model directly
tokenizer = AutoTokenizer.from_pretrained("Helsinki-NLP/opus-mt-fr-en")
model = AutoModelForSeq2SeqLM.from_pretrained("Helsinki-NLP/opus-mt-fr-en")

# Tokenize the input text
input_ids = tokenizer("Ce cours est produit par Hugging Face.", return_tensors="pt").input_ids

# Generate the translation
outputs = model.generate(input_ids)

# Decode the generated tokens back to text
result = tokenizer.decode(outputs[0], skip_special_tokens=True)
print(result)
```
